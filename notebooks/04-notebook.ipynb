```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt4o_llm = ChatOpenAI(model='gpt-4o', temperature=0)
response = gpt4o_llm.invoke("Hello, how are you?")
response.pretty_print()
```

```python
gpt3_llm = ChatOpenAI(model='gpt-3.5-turbo-0125', temperature=0)
response = gpt3_llm.invoke("Hello, how are you?")
response.pretty_print()
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in spanish")
msg_2 = HumanMessage(content="My name is Saul")
msg_3 = AIMessage(content="Hi Saul, how are you?")
msg_4 = HumanMessage(content="Whats my name?")

history = [msg_1,msg_2, msg_3, msg_4]
response = gpt3_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("gemini-2.5-flash",model_provider="google_genai",temperature=0)

response = llm.invoke(history)

response.pretty_print()

```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

gemini = ChatGoogleGenerativeAI(model='gemini-2.5-flash', temperature=0)
response = gemini.invoke(history)
response.pretty_print()
```
