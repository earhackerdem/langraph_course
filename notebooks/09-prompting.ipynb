```python
from langchain_core.prompts import PromptTemplate


template = """\
    # Instrucciones para el Asistente de IA
    
    Recuerda que hoy es {fecha_actual}
    
    ## Rol
    
    Eres un experto en marketing digital con 10 años de experiencia.
    
    ## Tarea
    Analiza el siguiente texto de un anuncio y proporciona 3 sugerencias para mejorarlo, enfocándote en el llamado a la acción
    
    ## Texto del Anuncio
    {texto_anuncio}
    
    ## Formato de salida
    - * Sugerencia 1: * [Tu primera sugerencia]
    - * Sugerencia 2: * [Tu segunda sugerencia]
    - * Sugerencia 3: * [Tu tercera sugerencia]
    """
    
prompt_template = PromptTemplate.from_template(template)
prompt = prompt_template.format(fecha_actual="2026-03-27", texto_anuncio="Esto es un anuncio de prueba")
print(prompt)
```

```python
import datetime
today = datetime.datetime.now().strftime("%Y-%m-%d")
prompt_template = PromptTemplate.from_template(template,partial_variables={"fecha_actual": today})
prompt = prompt_template.format(texto_anuncio="Esto es un anuncio de prueba")
print(prompt)
```
